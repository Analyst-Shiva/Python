```python
#Sample 1

import numpy as np
import matplotlib.pyplot as plt

width, height = 256, 256
image = np.zeros((height, width, 3), dtype=np.uint8)

square_size = 50
start_x = (width - square_size) // 2
start_y = (height - square_size) // 2
image[start_y:start_y + square_size, start_x:start_x + square_size, :] = [255, 255, 255]

plt.imshow(image)
plt.axis('off')
plt.show()
```

```python
#Sample 2

import numpy as np
import matplotlib.pyplot as plt

width, height, depth = 200, 100, 50
image_3d = np.zeros((height, width, depth), dtype=np.uint8)
image_3d[20:80, 50:150, 10:40] = 255

fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

nonzero_indices = np.nonzero(image_3d)
ax.scatter(nonzero_indices[1], nonzero_indices[0], nonzero_indices[2], c='r', marker='o')

ax.set_xlabel('X-axis')
ax.set_ylabel('Y-axis')
ax.set_zlabel('Z-axis')

plt.show()

```

```python
#Sample 2 axis 0
import numpy as np
import matplotlib.pyplot as plt


width, height = 200, 100
image = np.zeros((height, width), dtype=np.uint8)


image[:, :] = np.linspace(0, 255, width, dtype=np.uint8)


plt.imshow(image, cmap='gray', origin='lower', aspect='auto')
plt.colorbar(label='Intensity')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.show()
```

```python
#Sample 3
import numpy as np

array_3d = np.random.randint(0, 10, size=(3, 4, 5))

print("Original 3D array:")
print(array_3d)
print("\n")

print("Slices along axis 0:")
for i in range(array_3d.shape[0]):
    print(f"Slice {i} along axis 0:")
    print(array_3d[i, :, :])
    print("\n")

print("Slices along axis 1:")
for j in range(array_3d.shape[1]):
    print(f"Slice {j} along axis 1:")
    print(array_3d[:, j, :])
    print("\n")

print("Slices along axis 2:")
for k in range(array_3d.shape[2]):
    print(f"Slice {k} along axis 2:")
    print(array_3d[:, :, k])
    print("\n")
```

```python
#Sample 5

import numpy as np
import matplotlib.pyplot as plt

array_3d = np.random.randint(0, 255, size=(3, 4, 5), dtype=np.uint8)

fig, axes = plt.subplots(1, 3, figsize=(15, 5))

axes[0].imshow(array_3d.mean(axis=0), cmap='Reds')
axes[0].set_title('Slice along Axis 0')

axes[1].imshow(array_3d.mean(axis=1), cmap='Greens')
axes[1].set_title('Slice along Axis 1')

axes[2].imshow(array_3d.mean(axis=2), cmap='Oranges')
axes[2].set_title('Slice along Axis 2')

for ax in axes:
    ax.axis('off')

plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

vibgyor_gradient = np.array([
    [148, 0, 211],
    [75, 0, 130],
    [0, 0, 255],
    [0, 255, 0],
    [255, 255, 0],
    [255, 127, 0],
    [255, 0, 0]
])

width = 1000
height = 100
vibgyor_image = np.tile(vibgyor_gradient, (height, 1, 1))

plt.imshow(vibgyor_image)
plt.title('VIBGYOR Image')
plt.axis('off')
plt.show()

```

```python
#Sample 7

import numpy as np
import matplotlib.pyplot as plt

board_size = 8

chessboard = np.zeros((board_size, board_size, 3), dtype=np.uint8)

chessboard[1::2, ::2] = [255, 255, 255]
chessboard[::2, 1::2] = [255, 255, 255]

plt.imshow(chessboard)
plt.title('Chessboard')
plt.axis('off')
plt.show()
```
