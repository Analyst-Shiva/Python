```python
#3.Second row 2nd coloumn density plot with mean, median and mode, threshold lines, =1,-1 std
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

tips = sns.load_dataset('tips')
fig, axes = plt.subplots(1, 2, figsize=(12, 5))

sns.kdeplot(x='total_bill', data=tips, ax=axes[0])
axes[0].axvline(tips['total_bill'].mean(), color='red', ls='dashed', label='Mean')
axes[0].axvline(tips['total_bill'].median(), color='green', ls='dashed', label='Median')
axes[0].legend()

sns.kdeplot(x='total_bill', data=tips, ax=axes[1])
std_dev = tips['total_bill'].std()
axes[1].axvline(tips['total_bill'].mean() + std_dev, color='red', ls='dashed', label='+1 Std')
axes[1].axvline(tips['total_bill'].mean() - std_dev, color='red', ls='dashed', label='-1 Std')
axes[1].axvline(tips['total_bill'].mode().values[0], color='blue', ls='dashed', label='Mode')
axes[1].legend()

plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

tips = sns.load_dataset('tips')
fig, axes = plt.subplots(3, 1, figsize=(8, 15))

sns.kdeplot(x='total_bill', data=tips, ax=axes[0])
axes[0].axvline(tips['total_bill'].mean(), color='red', ls='dashed', label='Mean')
axes[0].axvline(tips['total_bill'].median(), color='green', ls='dashed', label='Median')
axes[0].legend()

# Other plots...

tip_size_counts = tips['size'].value_counts()
axes[2].pie(tip_size_counts, labels=tip_size_counts.index, autopct='%1.1f%%', startangle=90)
axes[2].set_aspect('equal')

plt.show()
```

```python
#6.Third row 2nd coloumn- count plot, text on the bar on the count bar
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

tips = sns.load_dataset('tips')
fig, axes = plt.subplots(3, 1, figsize=(8, 15))

sns.kdeplot(x='total_bill', data=tips, ax=axes[0])
axes[0].axvline(tips['total_bill'].mean(), color='red', ls='dashed', label='Mean')
axes[0].axvline(tips['total_bill'].median(), color='green', ls='dashed', label='Median')
axes[0].legend()

sns.countplot(x='day', data=tips, ax=axes[2])
for p in axes[2].patches:
    axes[2].annotate(f'{p.get_height()}', (p.get_x() + p.get_width() / 2., p.get_height()),
                     ha='center', va='center', xytext=(0, 10), textcoords='offset points')

tip_size_counts = tips['size'].value_counts()
axes[2].pie(tip_size_counts, labels=tip_size_counts.index, autopct='%1.1f%%', startangle=90)
axes[2].set_aspect('equal')

plt.show()
```
